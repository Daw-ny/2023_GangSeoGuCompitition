```python
# module
import folium
import pandas as pd
import numpy as np
```

```python
# load data
완료 = pd.read_csv('1차완료.csv')
횡단보도 = pd.read_csv('횡단보도_완료.csv', encoding = 'cp949')
```

```python
사용데이터 = pd.merge(횡단보도, 완료, on = 'crosswalk_id' ,how = 'left')[['crosswalk_id', 'lon', 'lat', '사고건수']]
```

```python
m = folium.Map(location=[37.55, 126.84], tiles='cartodbpositron', zoom_start=13)
```

```python
# Add  points to the map
for _, row in 사용데이터.iterrows():
    if row['사고건수'] > 30:
        folium.Circle(location = [row['lat'], row['lon']],
               radius = 8,
                popup = row['사고건수'],
                color = 'red'
        ).add_to(m)
    else:
        folium.Circle(location = [row['lat'], row['lon']],
               radius = 3,
                popup = row['사고건수'],
                color = 'gray'
        ).add_to(m)
```
